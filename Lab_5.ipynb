```python
class Main:
    @staticmethod
    def main():
        # creating scanner object for reading values
        
        str_expression = input("String expression: ")

        # calling the function
        result = Main.find_result(str_expression)
        print("\n\nResult:", result)

    @staticmethod
    def find_result(str_expression):
        # splitting the string 
        oper = str_expression.split(" ")

        # creating an instance of the Stack class
        stk = []

        # variables for temporary storing
        n1 = 0
        n2 = 0

        # loop through the inputted string
        for i in range(len(oper)):
            # checking if any operand 
            if oper[i] == "+":
                # delete top element from stack and store it in n1
                n1 = stk.pop()
                # delete the next top element and store it in n2
                n2 = stk.pop()
                # add the result to the top of the stack
                stk.append(n1 + n2)
            elif oper[i] == "-":
                n1 = stk.pop()
                n2 = stk.pop()
                stk.append(n1 - n2)
            elif oper[i] == "*":
                n1 = stk.pop()
                n2 = stk.pop()
                stk.append(n1 * n2)
            elif oper[i] == "/":
                n1 = stk.pop()
                n2 = stk.pop()
                stk.append(n1 / n2)
            else:
                
                num = int(oper[i])
                stk.append(num)


        
        # returning the result 
        return stk[-1]

# calling the main method to run the program
Main.main()



# teat Case 

stack = []

# Pushing elements onto the stack
stack.append(3)
print("stack:", stack)
# out put is stack: [3]

stack.append(5)
print("stack:", stack)
# out put is stack: [3, 5]

# Popping elements from the stack
n1 = stack.pop()
n2 = stack.pop()
stack.append(n1 + n2)
print("stack:", stack)
# out put is stack: [8]

stack.append(1)
print("stack:", stack)
# out put is stack: [8,1]

n1 = stack.pop()
n2 = stack.pop()
stack.append(n1 - n2)
print("stack:", stack)
# out put is stack: [-7]

# Printing the final result
result = stack[-1]
print("\nResult:", result)
# out put is Result: -7


```
