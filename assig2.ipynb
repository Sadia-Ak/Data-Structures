```python
class Main:
    @staticmethod
    def evaluate(expression):
        # the main class is define as evaluate 
        tokens = list(expression)
        # initialized as a list 
        # takes the method as expression as a input 
        values = []

        # store numbers 
        ops = []
        # used to store operators 
        i = 0
        # iterating through the tokens list 
        while i < len(tokens):
            
            if tokens[i] == ' ':
                i += 1
                continue

            
            if tokens[i].isdigit():
                sbuf = []
                
                # There may be more than one digits in number
                while i < len(tokens) and tokens[i].isdigit():
                    sbuf.append(tokens[i])
                    i += 1
                values.append(int(''.join(sbuf)))

                # i points to the character next to the digit,
                #the while loop also increases the i
                #  decrease the value of i by 1 
               
                i -= 1

            # Copening brace'
            elif tokens[i] == '(':
                ops.append(tokens[i])

            # Closing brace encountered 
            elif tokens[i] == ')':
                while ops[-1] != '(':
                    values.append(Main.apply_op(ops.pop(), values.pop(), values.pop()))
                ops.pop()

            
            elif tokens[i] in ['+', '-', '*', '/']:
                # Apply operator on top of 'ops'
                # to top two elements in values stack
                while ops and Main.has_precedence(tokens[i], ops[-1]):
                    values.append(Main.apply_op(ops.pop(), values.pop(), values.pop()))

                ops.append(tokens[i])  # Push current token to 'ops'.
            
            i += 1

        
        # ops to remaining values
        while ops:
            values.append(Main.apply_op(ops.pop(), values.pop(), values.pop()))

        return values[-1]  
    @staticmethod
    def has_precedence(op1, op2):
        if op2 in ['(', ')']:
            return False
        if (op1 in ['*', '/'] and op2 in ['+', '-']):
            return False
        else:
            return True

    @staticmethod
    def apply_op(op, b, a):
        if op == '+':
            return a + b
        if op == '-':
            return a - b
        if op == '*':
            return a * b
        if op == '/':
            if b == 0:
                raise ValueError("Can not be divide by zero")
            return a / b
        return 0

    @staticmethod
    def main():
        expression = input("Enter the expression: ")
        # call evaluate
        print("Evaluation =", Main.evaluate(expression))

# calling the main method 
Main.main()

# Test case 
expression = "10 * 2 - 15"
result = expression
print("Result:", result)
# out is Result: 10 * 2 - 15


```
